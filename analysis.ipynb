为poetry添加诗人名和简介绍

```python
poets_ser = poetrys_df.poet
ser_ls = []
for index,row in poetrys_df.iterrows():
    d = util.str2dict(poets_ser[index])
    ser_ls.append({"poetName":d["name"],"poetDesc":d.get("desc","暂无")})

poetrys_df = pd.concat([poetrys_df,pd.DataFrame(ser_ls)],axis=1)
```
